**Data Visualisations**

Bar Chats

```python
categorical_features = ['Displaced', 'Educational_special_needs', 'Debtor',
                         'Tuition_fees_up_to_date', 'Gender', 'Scholarship_holder', 'Age_at_enrollment',
                         'International']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 4, i)
    sns.countplot(x=feature, data=df, hue='Target')
    plt.title(f'A graph of {feature} by Target')

plt.tight_layout()
plt.show()
```

The above Histograms shows how differmt factors affect the Success of a student. For example Age at enrollment, shows that the higher aged students tend to dropout due to different life factors and hindernaces, Gender ie it shows that the 2 genders perfom differently, Etc
